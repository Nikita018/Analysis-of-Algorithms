```
#interval_list = range(-10000,10000)

interval_dict={}
for idx in range(-10000,10001):
  interval_dict[idx]=[]

print(len(interval_dict.keys()))
```

```
count=0
for i in list(input_dict):
  for j in list(interval_dict):
    second_half = str(int(j)-int(i))
    if(second_half in input_dict.keys() and second_half != i):
      del interval_dict[j]
      count=count + 1
      print("count",count)
  del input_dict[i]
```
